## 1.
The time required for servicing transmissions is normally distributed with mean = 45 minutes and std dev = 8 minutes. The service manager plans to have work begin on the transmission of a customer’s car 10 minutes after the car is dropped off and the customer is told that the car will be ready within 1 hour from drop-off. What is the probability that the service manager cannot meet his commitment?


```python
# Given
mu1 = 45
sigma1 = 8
x1 = 60-10
```

```python
# Probability that servicing time > 60 min
from scipy import stats
p=(1-stats.norm.cdf(x1,mu1,sigma1))
p
```

```python
print(f'probability that the service manager cannot meet his commitment: {p:.3f}')
```

##2.
The current age (in years) of 400 clerical employees at an insurance claims processing center is normally distributed with mean  = 38 and Standard deviation  =6. For each statement below, please specify True/False. If false, briefly explain why.

A.	More employees at the processing center are older than 44 than between 38 and 44.

B.	A training program for employees under the age of 30 at the center would be expected to attract about 36 employees.


```python
# Given
mu2 = 38
sigma2 = 6
n2=400
```

### calculation for
A-More employees at the processing center are older than 44 than between 38 and 44.

```python
# calculation for number of employees at the processing center are older than 44
e_44=(1-stats.norm.cdf(44,mu2,sigma2))*n2
print('number of employees older than 44 at the processing center are =', e_44.astype('int'))
```

```python
# calculation for number of employees between 38 and 44 at the processing center
e_38_44 = (stats.norm.cdf(44,mu2,sigma2)-stats.norm.cdf(38,mu2,sigma2))*n2
print('number of employees between 38 and 44 at the processing center are =', e_38_44.astype('int'))
```

```python
if e_44>e_38_44:
  print('More employees at the processing center are older than 44 than between 38 and 44')
else:
  print('More employees at the processing center are between 38 and 44 than older than 44')

```

### calculation for
B. A training program for employees under the age of 30 at the center would be expected to attract about 36 employees.

```python
# Calculation for number of employees under the age of 30 at the center
e_30 = (stats.norm.cdf(30,mu2,sigma2))*n2
print('number of employees under the age of 30 at the processing center are =', e_30.astype('int'))
```

#4.
Let X ~ N(100, 20^2). Find two values, a and b, symmetric about the mean, such that the probability of the random variable taking a value between them is 0.99.

A.	90.5, 105.9

B.	80.2, 119.8

C.	22, 78

D.	48.5, 151.5

E.	90.1, 109.9
   

```python
# Given
mu4 = 100
sigma4 = 20
x4 = [90.5,80.2,22,48.5,90.1] ## as two values, a and b, symmetric about the mean, so taking only a from each option
```

```python
for i in x4:
  p=(1-2*stats.norm.cdf(i,mu4,sigma4))
  print('The probability for option {} is: {}'.format(i,p))
  if p*100>99:
        print(f'option containing {i} is correct')
```

Consider a company that has two different divisions. The annual profits from the two divisions are independent and have distributions Profit1 ~ N(5, 3^2) and Profit2 ~ N(7, 4^2) respectively. Both the profits are in $ Million.

Answer the following questions about the total profit of the company in Rupees. Assume that $1 = Rs. 45

A.	Specify a Rupee range (centered on the mean) such that it contains 95% probability for the annual profit of the company.

C.	Which of the two divisions has a larger probability of making a loss in a given year?


```python
# Given
mu5_1=5
mu5_2=7
variance_1=3**2
variance_2=4**2
```

```python
## before proceeding normalise both divisions into 1
mu5=mu5_1+mu5_2
variance= variance_1 + variance_2

import numpy as np
sigma5=np.sqrt(variance)
```

### A.
Specify a Rupee range (centered on the mean) such that it contains 95% probability for the annual profit of the company.

```python
probability = 0.95
z_score = stats.norm.ppf(probability)
z_score
```

```python
# z_score = (x-mu)/sigma
x5_1=mu5 + (z_score*sigma5)
x5_2=mu5 - (z_score*sigma5)
```

```python
print('Rupee range (centered on the mean) such that it contains 95% probability\
 for the annual profit of the company is between\n {}, and {}'.format(x5_2*45*10**6,x5_1*45*10**6))
```

Specify the 5th percentile of profit (in Rupees) for the company

```python
# X such that the cumulative probability to the left of X is 0.05 (5th percentile)
```

```python
probability2 = 0.05
z_score2 = -stats.norm.ppf(probability) ## -ve sign for less than 50% proobability
z_score2
```

```python
x5_b=mu5 + (z_score2*sigma5)
```

```python
print('5th percentile of profit (in Rupees) for the company\n {}'.format(x5_b*45*10**6))
```

#8.
How many randomly selected employers (minimum number) must we contact in order to guarantee a margin of error of no more than 4% (at 95% confidence)?

A.	600
B.	400
C.	550
D.	1000

```python
# Given
Margin_of_error8 = 0.04

```

```python
## test_statistics
### Z-test
alpha8=0.05/2 # 95% confidence
Z_statistic8 = stats.norm.ppf(1-alpha8)
Z_statistic8
```

```python
# Margin_of_error = Z_statistic * sigma /sqrt(n)
## since sigma is not given => sigma^2 = p(1-p)  ## taking binomial distribution closer to normal distribution
p= 0.5
n8= p*(1-p)*(Z_statistic8/Margin_of_error8)**2
print(f'employers (minimum number) must we contact in order to guarantee a margin of error: {int(n8)}')
```

so answer A is correct

#3.
Auditors at a small community bank randomly sample 100 withdrawal transactions made during the week at an ATM machine located near the bank’s main branch. Over the past 2 years, the average withdrawal amount has been $50 with a standard deviation of $40. Since audit investigations are typically expensive, the auditors decide to not initiate further investigations if the mean transaction amount of the sample is between $45 and $55. What is the probability that in any given week, there will be an investigation?

A.	1.25%
B.	2.5%
C.	10.55%
D.	21.1%
E.	50%


```python
# Given
mu = 50
sigma = 40
n = 100
```

```python
# Z_score = (x-mu)/(sigma/np.sqrt(n))
x1=45
x2=55
Z_score1 = (x1-mu)/(sigma/np.sqrt(n))
Z_score2 = (x2-mu)/(sigma/np.sqrt(n))
Z_score1, Z_score2
```

```python
# probability that mean transaction amount of the sample is between  45 and 55.
p=stats.norm.cdf(Z_score2)-stats.norm.cdf(Z_score1)

# probability that in any given week, there will be an investigation
# when mean transaction amount of the sample is not between  45 and 55.
print('probability that in any given week, there will be an investigation', (1-p)*100)
```

so, answer is D
